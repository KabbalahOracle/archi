```python
ner_data['ent'] = ner_data['ent'].apply(lambda x: revise_ent(x, 'ATTR', 'AEC'))
```

```python
def revise_ent(ents, old_label, new_label):
    for idx, ent in enumerate(ents):
        if ent[2] == old_label:
            ents[idx] = (ent[0], ent[1], new_label)
    return ents
```
